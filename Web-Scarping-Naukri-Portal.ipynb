```python
import pandas as pd
from bs4 import BeautifulSoup
from datetime import date 
```

```python
today=str(date.today())
```

```python
from selenium import webdriver
options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
options.add_argument('--headless')
driver = webdriver.Chrome("chromedriver.exe", chrome_options=options)
```

```python
data=pd.read_csv("link_by_areas.csv")
```

```python
pd.set_option("display.max_colwidth",-1)
```

```python
data
```

```python
final_data=pd.DataFrame()
```

```python
def scrap(link):
    driver.get(link)
    print(link)
    html = driver.page_source
    soup = BeautifulSoup(html)
    job_post=soup.find_all("article",{"class":"jobTuple bgWhite br4 mb-8"})
    return job_post
```

```python
for col in data.iterrows():
    area=col[1].type
    job_post=None
    while not job_post:
        job_post=scrap(col[1].link)
    for job in job_post:
        global final_data
        title=""
        company=""
        salary=""
        exp=""
        location=""
        description=""
        date=""
        tags=""
        if job.find("a", {"class": "title fw500 ellipsis"}):
            title=job.find("a", {"class": "title fw500 ellipsis"}).text
        if job.find("a", {"class": "subTitle ellipsis fleft"}):
            company = job.find("a", {"class": "subTitle ellipsis fleft"}).text
        if job.find("li", {"class": "fleft grey-text br2 placeHolderLi experience"}):
            exp=job.find("li", {"class": "fleft grey-text br2 placeHolderLi experience"}).text
        if job.find("li", {"class": "fleft grey-text br2 placeHolderLi salary"}):
            salary=job.find("li", {"class": "fleft grey-text br2 placeHolderLi salary"}).text
        if job.find("li", {"class": "fleft grey-text br2 placeHolderLi location"}):
            location=job.find("li", {"class": "fleft grey-text br2 placeHolderLi location"}).text
        if job.find("div", {"class": "job-description fs12 grey-text"}):
            description=job.find("div", {"class": "job-description fs12 grey-text"}).text
        tags=job.find("ul", {"class": "tags has-description"})
        da=job.find("div", {"class": "type br2 fleft grey"})
        if(da!=None):
            date=da.find("span", {"class": "fleft fw500"}).text
        elif(job.find("div", {"class": "type br2 fleft green"})):
            da=job.find("div", {"class": "type br2 fleft green"})
            date=da.find("span", {"class": "fleft fw500"}).text
        dic={"Job Title":title,"Company":company,"Experience":exp,"Salary":salary,"Location":location,
             "Description":description,"Tags":tags,"Function Area":area,"Posting Date":date,"Scraping Date":today}
        final_data=final_data.append(dic,ignore_index=True)
```

```python
final_data.to_csv("scraping_data.csv.csv",index=False)
```
